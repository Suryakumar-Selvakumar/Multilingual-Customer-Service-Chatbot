```python
# Import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import random
import time
import os
import nltk
import numpy as np
from nltk.tokenize import word_tokenize
from nltk.translate import meteor_score
from rouge_score import rouge_scorer
import sacrebleu
from comet.models import download_model, load_from_checkpoint
import seaborn as sns
from math import pi
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
import warnings
warnings.filterwarnings("ignore")
```

### Load Final Model

```python
# Load the model and tokenizer from the directory
model_dir = "models/final_mbart_model"
model = MBartForConditionalGeneration.from_pretrained(model_dir).to('cuda')  # Move model to GPU
tokenizer = MBart50TokenizerFast.from_pretrained(model_dir)

# Function to generate a response
def generate_response(input_text, source_lang, target_lang, max_length=256, num_beams=5, min_length=125):
    # Set the source language token
    tokenizer.src_lang = source_lang
    
    # Tokenize and move the inputs to the GPU
    model_inputs = tokenizer(input_text, return_tensors="pt", padding=True)
    model_inputs = {key: value.to('cuda') for key, value in model_inputs.items()}  # Move all inputs to GPU
    
    # Ensure the model and its inputs/outputs are on the same device
    generated_tokens = model.generate(
        input_ids=model_inputs['input_ids'],
        attention_mask=model_inputs['attention_mask'],
        forced_bos_token_id=tokenizer.lang_code_to_id[target_lang],
        eos_token_id=tokenizer.eos_token_id,  # Add end-of-sequence token to stop generation early
        min_length=min_length,
        max_length=max_length,
        num_beams=num_beams,
        no_repeat_ngram_size=2,
        early_stopping=True
    )

    # Decode and return the response
    return tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
```

### Sanity Check with a Few Samples

```python
# Test the function with some sample inputs and language-specific lengths
inputs = {
    "en": "How can I cancel my order?",
    "fr": "Comment puis-je annuler ma commande?",
    "es": "Cmo puedo cancelar mi pedido?"
}

# Generating responses with language-specific lengths
for lang, input_text in inputs.items():
    response = generate_response(input_text, source_lang=f'{lang}_XX', target_lang=f'{lang}_XX', max_length=627, min_length=125)  
    print(f"Input: {input_text} | Response: {response}")
    print("\n")
```

```python
# Function to clean up special tokens from the dataset
def clean_special_tokens(text):
    # Remove any special tokens like <s>, </s>, or language codes
    cleaned_text = text.replace('</s>', '').replace('en_XX', '').replace('fr_XX', '').replace('es_XX', '')
    return cleaned_text.strip()

# Load the multilingual dataset
multilingual_df = pd.read_csv('data/Multilingual_Customer_Support_Training_Dataset.csv')

# Clean the dataset's instructions and responses
multilingual_df['instruction'] = multilingual_df['instruction'].apply(clean_special_tokens)
multilingual_df['response'] = multilingual_df['response'].apply(clean_special_tokens)

# Save the cleaned dataset to a new CSV file
cleaned_dataset_path = 'data/Multilingual_Customer_Support_Training_Dataset_Cleaned.csv'
multilingual_df.to_csv(cleaned_dataset_path, index=False)

print(f"Cleaned dataset saved to {cleaned_dataset_path}")
```

```python
# Randomly sample a few rows from the dataset to sanity check
sample_df = multilingual_df.sample(n=5, random_state=42)

# Iterate over the samples and generate model responses
for idx, row in sample_df.iterrows():
    lang_code = row['language'][:2]  # Get language prefix like 'en', 'fr', 'es'
    instruction = row['instruction']
    
    response = generate_response(input_text=instruction, source_lang=f'{lang_code}_XX', target_lang=f'{lang_code}_XX', max_length=627)
    print(f"Input: {instruction} | Response: {response}\n")
```

### Initialization

```python
# Set seeds for reproducibility
random.seed(42)

# Set up nltk data path
nltk_data_path = 'data/nltk_data'
nltk.data.path.append(nltk_data_path)

# Ensure punkt is available for tokenization
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab', download_dir=nltk_data_path)
```

### Normalize response data for Evaluation

```python
# Function to normalize text by removing line breaks and extra spaces
def normalize_text(text):
    # Replace multiple spaces with a single space and strip the text
    return ' '.join(text.replace('\n', ' ').split()).strip()

# Apply normalization to responses in the dataset
multilingual_df['normalized_response'] = multilingual_df['response'].apply(normalize_text)
```

### Generate Model responses for a Sample Subset of Instructions

```python
# File path to save generated responses, references, and response times
combined_file = "data/generated_responses_combined.csv"

# Check if the generated responses are already saved
if not os.path.exists(combined_file):
    # 20% subset for testing
    subset_size = int(0.2 * len(multilingual_df)) 
    subset_df = multilingual_df.sample(n=subset_size, random_state=42)

    # Generate responses for the subset
    data = []

    start_time = time.time()  # Start the timer
    counter = 1  # Initialize a counter for tracking

    for idx, row in subset_df.iterrows():
        lang_code = row['language'][:2] 
        instruction = row['instruction']

        # Start time for generation
        generation_start = time.time()

        # Generate response
        response = generate_response(input_text=instruction, source_lang=f'{lang_code}_XX', target_lang=f'{lang_code}_XX', max_length=627)

        # End time for generation
        generation_time = time.time() - generation_start

        # Normalize both generated response and reference response
        normalized_generated_response = normalize_text(response)
        normalized_reference = normalize_text(row['normalized_response'])

        # Append the row data
        data.append({
            "instruction": instruction,
            "generated_response": normalized_generated_response,
            "reference": normalized_reference,
            "response_time": generation_time
        })

        # Print the time taken for this specific generation, with normal counter
        print(f"Sample: {counter}/{subset_size}\tTime taken for Sample {idx}: {generation_time:.2f} seconds")

        # Increment the counter
        counter += 1

    # Convert the data to a DataFrame and save it to a CSV file
    combined_df = pd.DataFrame(data)
    combined_df.to_csv(combined_file, index=False)

    print(f"Generated responses, references, and response times saved to {combined_file}")

else:
    # Load the previously saved generated responses, references, and response times
    combined_df = pd.read_csv(combined_file)
    print("Loaded saved instructions, generated responses, references, and response times.")

# Extract columns for testing from the loaded or newly created DataFrame
generated_responses = combined_df['generated_response'].tolist()
references = combined_df['reference'].tolist()
response_times = combined_df['response_time'].tolist()
instructions = combined_df['instruction'].tolist()
```

### BLEU Score Calculation

```python
# Compute BLEU Scores using sacrebleu
bleu = sacrebleu.corpus_bleu(generated_responses, [[ref] for ref in references])
precisions = bleu.precisions

print("\nBLEU Scores:")
print(f"BLEU-1 Score: {precisions[0]:.2f}")
print(f"BLEU-2 Score: {precisions[1]:.2f}")
print(f"BLEU-3 Score: {precisions[2]:.2f}")
print(f"BLEU-4 Score: {precisions[3]:.2f}")
```

```python
# Extract BLEU precision scores
bleu_scores = [precisions[0], precisions[1], precisions[2], precisions[3]]
bleu_labels = ['BLEU-1', 'BLEU-2', 'BLEU-3', 'BLEU-4']

# Plotting BLEU scores
plt.figure(figsize=(10, 6))
sns.barplot(x=bleu_labels, y=bleu_scores, palette='viridis')
plt.ylim(0, 100)
plt.ylabel('BLEU Score (%)')
plt.title('BLEU Score Breakdown')
plt.show()
```

#### BLEU Scores over Different Sentence Lengths

```python
# Add a length column to combined_df
combined_df['length'] = combined_df['generated_response'].apply(len)

# Set bins based on the actual data and ensure they are strictly increasing
max_length = combined_df['length'].max()
bins = [0, 250, 500, max(501, max_length + 1)]
labels = ['Short', 'Medium', 'Long']

# Assign length categories based on the new bins
combined_df['length_category'] = pd.cut(combined_df['length'], bins=bins, labels=labels, include_lowest=True)

# Calculate BLEU scores for each length category, ensuring non-empty groups
bleu_scores_by_length = []
categories = []

for category, group in combined_df.groupby('length_category'):
    if len(group) > 0:
        # Calculate BLEU score for non-empty group
        bleu_score = sacrebleu.corpus_bleu(group['generated_response'].tolist(), [[ref] for ref in group['reference']])
        bleu_scores_by_length.append(bleu_score.precisions[3])
        categories.append(category)

# Plotting BLEU-4 scores by length category
plt.figure(figsize=(10, 6))
sns.barplot(x=categories, y=bleu_scores_by_length, palette='autumn')
plt.ylim(0, 100)
plt.ylabel('BLEU-4 Score (%)')
plt.title('BLEU-4 Score by Sentence Length Category')
plt.show()
```

### ROUGE Score Calculation

```python
# Compute ROUGE scores using rouge-score library
rouge_scorer_instance = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
rouge1_scores, rouge2_scores, rougeL_scores = [], [], []

for gen_resp, ref_resp in zip(generated_responses, references):
    scores = rouge_scorer_instance.score(ref_resp, gen_resp)
    rouge1_scores.append(scores['rouge1'].fmeasure)
    rouge2_scores.append(scores['rouge2'].fmeasure)
    rougeL_scores.append(scores['rougeL'].fmeasure)

avg_rouge1 = sum(rouge1_scores) / len(rouge1_scores)
avg_rouge2 = sum(rouge2_scores) / len(rouge2_scores)
avg_rougeL = sum(rougeL_scores) / len(rougeL_scores)

print("\nROUGE Scores:")
print(f"ROUGE-1 Score: {avg_rouge1:.2f}")
print(f"ROUGE-2 Score: {avg_rouge2:.2f}")
print(f"ROUGE-L Score: {avg_rougeL:.2f}")
```

```python
# ROUGE Scores
rouge_scores = [avg_rouge1, avg_rouge2, avg_rougeL]
rouge_labels = ['ROUGE-1', 'ROUGE-2', 'ROUGE-L']

# Plotting ROUGE scores
plt.figure(figsize=(10, 6))
sns.barplot(x=rouge_labels, y=rouge_scores, palette='magma')
plt.ylim(0, 1)
plt.ylabel('ROUGE Score (F1)')
plt.title('ROUGE Score Breakdown')
plt.show()
```

### METEOR Score Calculation

```python
# Compute METEOR scores using NLTK
meteor_scores = []

for gen_resp, ref_resp in zip(generated_responses, references):
    # Tokenize response and reference for METEOR calculation
    tokenized_generated_response = word_tokenize(gen_resp)
    tokenized_reference = word_tokenize(ref_resp)
    meteor_scores.append(meteor_score.meteor_score([tokenized_reference], tokenized_generated_response))

avg_meteor = sum(meteor_scores) / len(meteor_scores)
print(f"\nMETEOR Score: {avg_meteor:.2f}")
```

```python
# Plotting METEOR score
plt.figure(figsize=(8, 6))
plt.bar(['METEOR'], [avg_meteor], color='coral')
plt.ylim(0, 1)
plt.ylabel('METEOR Score')
plt.title('METEOR Score for Generated Responses')
plt.show()
```

### Response Time Evaluation

```python
# Calculate 95th percentile response time
percentile_95_response_time = np.percentile(response_times, 95)

# Calculate total and average generation time
total_time = sum(response_times)
average_time_per_response = total_time / len(generated_responses)

print("\nLatency Metrics:")
print(f"Total time for generating {len(generated_responses)} responses: {total_time:.2f} seconds")
print(f"Average time per response: {average_time_per_response:.2f} seconds")
print(f"95th Percentile Response Time: {percentile_95_response_time:.2f} seconds")
```

```python
# Plot histogram for response times
plt.figure(figsize=(10, 6))
sns.histplot(response_times, bins=30, kde=True, color='blue')
plt.xlabel('Response Time (seconds)')
plt.ylabel('Frequency')
plt.title('Distribution of Response Times for Generated Responses')
plt.show()

# Plot boxplot to visualize percentiles
plt.figure(figsize=(10, 4))
sns.boxplot(response_times, color='skyblue')
plt.xlabel('Response Time (seconds)')
plt.title('Boxplot of Response Times')
plt.show()
```

### TER Score Calculation

```python
# Compute TER scores using sacrebleu
ter = sacrebleu.corpus_ter(generated_responses, [[ref] for ref in references])
avg_ter = ter.score / 100  # TER score is a percentage, convert it to decimal

print(f"\nTER Score: {avg_ter:.2f}")
```

```python
# Plotting TER score
plt.figure(figsize=(8, 6))
plt.bar(['TER'], [avg_ter], color='limegreen')
plt.ylim(0, 1)
plt.ylabel('TER Score')
plt.title('Translation Edit Rate (TER) for Generated Responses')
plt.show()
```

### CHaRacter-level F-scores Calculation

```python
# Compute ChrF3 scores
chrf3 = sacrebleu.corpus_chrf(generated_responses, [[ref] for ref in references], beta=3)
print(f"\nChrF3 Score: {chrf3.score:.2f}")

# Compute normal ChrF scores
chrf = sacrebleu.corpus_chrf(generated_responses, [[ref] for ref in references])
print(f"\nChrF Score: {chrf.score:.2f}")

# Compute ChrF++ scores
chrf_plus = sacrebleu.corpus_chrf(generated_responses, [[ref] for ref in references], word_order=2)
print(f"\nChrF++ Score: {chrf_plus.score:.2f}")
```

```python
# Character-level F-score values
chrf_scores = [chrf.score, chrf3.score, chrf_plus.score]
chrf_labels = ['ChrF', 'ChrF3', 'ChrF++']

# Plotting ChrF scores
plt.figure(figsize=(10, 6))
sns.barplot(x=chrf_labels, y=chrf_scores, palette='cool')
plt.ylim(0, 100)
plt.ylabel('ChrF Score (%)')
plt.title('Character-Level F-Score Breakdown (ChrF, ChrF3, ChrF++)')
plt.show()
```

### COMET Score Calculation

```python
# Load the COMET model
model_path = "models/comet/models--Unbabel--wmt20-comet-qe-da/snapshots/2e7ffc84fb67d99cf92506611766463bb9230cfb/checkpoints/model.ckpt"
comet_model = load_from_checkpoint(model_path)

# Prepare data for COMET prediction
data = [
    {"src": src, "mt": mt, "ref": ref}
    for src, mt, ref in zip(instructions, generated_responses, references)
]

# Predict COMET scores for the data
comet_output = comet_model.predict(data, batch_size=8, gpus=1) 

# Extract scores and system_score from comet_output
comet_scores = comet_output.scores  # List of individual scores
system_comet_score = comet_output.system_score  # System-level score for the entire data

# Calculate average COMET score from individual scores
avg_comet = sum(comet_scores) / len(comet_scores)
print(f"\nAverage COMET Score (from individual scores): {avg_comet:.2f}")
print(f"System-level COMET Score: {system_comet_score:.2f}")
```

```python
# Individual vs System COMET Score
comet_labels = ['Average COMET Score', 'System-Level COMET Score']
comet_scores = [avg_comet, system_comet_score]

# Plotting COMET scores
plt.figure(figsize=(10, 6))
sns.barplot(x=comet_labels, y=comet_scores, palette='spring')
plt.ylim(0, 1)
plt.ylabel('COMET Score')
plt.title('COMET Scores for Generated Responses')
plt.show()
```

### Overall Performance

```python
#Normalizing the metrics for the plot
metrics = ['BLEU-4', 'ROUGE-L', 'METEOR', 'ChrF3', 'TER']
max_values = [100, 1, 1, 100, 1]  # Max values for each metric
values = [
    precisions[3] / max_values[0],  # BLEU-4
    avg_rougeL / max_values[1],     # ROUGE-L
    avg_meteor / max_values[2],     # METEOR
    chrf3.score / max_values[3],    # ChrF3
    avg_ter / max_values[4]         # TER
]
values += values[:1]  # Closing the radar chart loop

# Radar chart setup
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Angles for radar chart
angles = [n / float(len(metrics)) * 2 * pi for n in range(len(metrics))]
angles += angles[:1]  # Closing the radar chart loop

# Plot data
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.4)

# Set categories and labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(metrics)

plt.title('Overall Performance Metrics Summary', pad=30) 
plt.show()
```

```python
# List of language codes to evaluate
languages = ['en', 'fr', 'es']
language_codes = {
    'en': 'en_XX',
    'fr': 'fr_XX',
    'es': 'es_XX'
}

# Iterate through each language and save the necessary data to CSV files for evaluation
data_save_path = "data/language_specific/"
os.makedirs(data_save_path, exist_ok=True)

for lang in languages:
    csv_file_path = f"{data_save_path}generated_responses_{lang}.csv"
    
    if not os.path.exists(csv_file_path):
        print(f"\n--- Generating Responses and Latencies for Language: {lang} ---\n")
        
        # Filter the dataset for the specific language - 20% data from each language
        language_sample_df = multilingual_df[multilingual_df['language'] == lang].sample(n=5374, random_state=42)

        # Store latencies, generated responses, and reference responses for the current language
        latencies = []
        generated_responses = []
        references = []

        # Initialize a counter for tracking the sample number
        counter = 1
        subset_size = len(language_sample_df)

        # Iterate over the sampled rows to measure latency and collect responses
        for idx, row in language_sample_df.iterrows():
            instruction = row['instruction']

            # Record the start time
            start_time = time.time()

            # Generate response
            response = generate_response(input_text=instruction, source_lang=language_codes[lang], target_lang=language_codes[lang], max_length=627)

            # Record the end time and calculate latency
            latency = time.time() - start_time
            latencies.append(latency)

            # Store generated response and reference response
            generated_responses.append(normalize_text(response))
            references.append(normalize_text(row['response']))

            # Print the latency for this specific generation, with the updated counter format
            print(f"Sample: {counter}/{subset_size}\tTime taken for Sample {idx}: {latency:.2f} seconds")

            # Increment the counter
            counter += 1

        # Save all necessary data to a CSV file for each language
        data = {
            "instruction": language_sample_df['instruction'].tolist(),
            "generated_response": generated_responses,
            "reference": references,
            "latency": latencies
        }
        language_data_df = pd.DataFrame(data)
        language_data_df.to_csv(csv_file_path, index=False)
        print(f"Data for language {lang} saved to {csv_file_path}")

    else:
        print(f"Data for language {lang} already exists at {csv_file_path}. Skipping generation.")
```
